```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
time_space = np.linspace(1,50,num=50)

u = 8
state_space = 70

o_c_cai = state_space*(u**time_space)
o_c_si = (state_space*u)**time_space
o_c_dpefe = state_space*u*time_space
o_c_postdiction = state_space + time_space*0
```

```python
plt.plot(time_space, o_c_cai)
plt.plot(time_space, o_c_si)
plt.plot(time_space, o_c_dpefe)
# plt.plot(time_space, o_c_postdiction)

plt.yscale("log")
plt.xlabel("Time horizon of planning")
plt.ylabel("$ \mathcal{O}$(Complexity)")
plt.legend(["Classical Active inference", "Sophisticated Inference", "DP-EFE method"])
plt.title("Comparison of computational complexity")
plt.savefig('Complexity3Methods.eps', dpi=500, bbox_inches='tight', format='eps')
```

```python
plt.plot(time_space, o_c_dpefe)
plt.plot(time_space, o_c_postdiction)
plt.xlabel("Time horizon of planning")
plt.ylabel("$ \mathcal{O}$(Complexity)")
plt.legend(["DPEFE method", "CL Method (No planning)"])
plt.title("Comparison of computational complexity")
plt.savefig('ComplexityDPEFEPostDic.png', dpi=500, bbox_inches='tight', format='png')
```
