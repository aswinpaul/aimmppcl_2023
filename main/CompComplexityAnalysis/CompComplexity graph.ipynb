```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
time_space = np.linspace(1,50,num=50)

u = 8
state_space = 70

o_c_cai = state_space*(u**time_space)
o_c_si = (state_space*u)**time_space
o_c_dpefe = state_space*u*time_space
o_c_postdiction = state_space #+ time_space*0
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

# ax.plot(time_space, o_c_cai, label = "Classical Active inference")
# ax.plot(time_space, o_c_si, label = "Sophisticated Inference")
# ax.plot(time_space, o_c_dpefe, label = "DPEFE method")

ax.plot(time_space, o_c_cai)
ax.plot(time_space, o_c_si)
ax.plot(time_space, o_c_dpefe)

highlight = [1, o_c_postdiction]
ax.scatter(highlight[0], highlight[1] , marker='v', color='r', label = "CL method (T = 1)")

ax.legend(loc = (0.025,0.4))
plt.yscale("log")
plt.xlabel("Time horizon of planning")
plt.ylabel("$\mathcal{O}$ of computational complexity (log scale)")
plt.title("Computational complexity vs $T$ [Fixed $Card(S)$ and $Card(U)$]")
plt.savefig('Complexity4Methods.png', dpi=500, bbox_inches='tight', format='png')
```
